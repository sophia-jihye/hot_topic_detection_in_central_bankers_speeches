```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
data_filepath = os.path.join('./data/Fed_chairs_199701-201909_491.csv')
save_dir = 'Fed_Speech_Transcripts'
if not os.path.exists(save_dir): os.makedirs(save_dir)
```

```python
df = pd.read_csv(data_filepath)
df.head()
```

```python
df['url'] = df['key'].apply(lambda x: 'https://www.bis.org/review/{}.htm'.format(x))
df.head()
```

```python
title_format = 'UploadedOn={}_Speaker={}'
content_format = 'URL={}\nPDF_URL={}\nTITLE={}\nSHORT_INFO={}\n\n{}'

for _, row in tqdm(df.iterrows()):
    title = title_format.format(row['date'], row['speaker'])
    content = content_format.format(row['url'], row['pdf_url'], row['title'], row['short_info'], row['content'])

    with open(os.path.join(save_dir, '{}.txt'.format(title)), 'w', encoding='utf-8') as f:
        f.write(content)
```
